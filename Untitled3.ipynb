```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
#Modelo KNN -K-Vecinos mas cercanos para clasificaciÃ³n
#Cargamos los datos
clientes = pd.read_csv("creditos.csv")
clientes

#Separamos los clientes que si pagaron el credito, y los que no
buenos = clientes[clientes["cumplio"]==1]
malos = clientes[clientes["cumplio"]==0]

```

```python
#Podemos hacer una grafica de los que pagan y de los que no, en funciÃ³n de su edad y la cantidad del credito
plt.scatter(buenos["edad"], buenos["credito"], color="green")

plt.scatter(malos["edad"], malos["credito"], color="red")

plt.ylabel("Cantidad del credito")
plt.xlabel("Edad")
plt.show()
```

```python
#Preparamos los datos (los escalamos)
#transformamos los datos para que se equivalen, de 0 a 1
datos = clientes[["edad", "credito"]]
clase = clientes["cumplio"]

escalador = preprocessing.MinMaxScaler()
datos = escalador.fit_transform(datos)
datos
```

```python
#Creamos el modelo clasificador KNN
#La variable n_neighbors son el numero de vecinos que se toman en cuenta para crear el modelo, utilizaremos uno bÃ¡sico,
#se suele usar la raiz cuadrada del numero de instancias, en este caso serÃ­a la raiz cuadrada de 200
clasificador = KNeighborsClassifier(n_neighbors=3)
clasificador.fit(datos, clase)
```

```python
#AÃ±adimos un nuevo cliente, para el modelo
edad = 53
cantidad = 350000

#Escalamos los datos del nuevo solicitante
solicitante = escalador.transform([[edad, cantidad]])

#Podemos ver la clase y la probabilidad que tiene el solicitante en ser de la clase buena o mala
print("Clase", clasificador.predict(solicitante))
print("Probabilidades por clase", clasificador.predict_proba(solicitante))
```

```python
#Podemos visualizar el nuevo solicitante
print("Clase", clasificador.predict(solicitante))
print("Probabilidades por clase", clasificador.predict_proba(solicitante))
plt.scatter(buenos["edad"], buenos["credito"], color="green")

plt.scatter(malos["edad"], malos["credito"], color="red")

plt.scatter(edad, cantidad, color="black")
plt.ylabel("Cantidad del credito")
plt.xlabel("Edad")
plt.show()
```
