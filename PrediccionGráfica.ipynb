```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1990, 1995, 2000, 2005, 2010, 2015, 2020])
y = np.array([120.50, 150.25, 190.75, 230.80, 260.40, 280.60, 300.00])
y = np.flip(y)
anno = 2025
y_pred_vec = np.array([])
for i in range(10):
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, anno)
    print(f"Para grado {i} la predicci칩n es {p}")

    #Calculamos el error medio cuadratico para saber que funcion tiene menor porcentaje de error
    y_pred_vec = np.array([])
    for j in x:
      y_pred = np.polyval(coef, j)
      y_pred_vec = np.append(y_pred_vec, y_pred)

    EMC = (sum( (y - y_pred_vec)**2))/len(y)
    print(f"Para grado {i} el EMC es: {EMC}")


    x1 = np.linspace(1990, anno + 1, 1000)
    y1 = np.polyval(coef, x1)
    plt.figure(figsize=[20,10])
    plt.title("Beneficios a lo largo de los a침os")
    plt.scatter(x,y,s=120,c='blueviolet')
    plt.plot(x1,y1,"--",linewidth=3,color='orange')
    plt.scatter(anno,p,s=200,c='red')
    plt.yticks(range(100,400,20))
    plt.xticks(np.arange(1990, anno + 1, 1))  # Establecer los ticks del eje x
    plt.grid(True)
    ax = plt.gca()
    ax.set_xlabel("$a침os$")
    ax.set_ylabel("Beneficios")
    plt.show()
```

```python
todoslosgrados = np.array([])
emcdecadagrado = np.array([])

anno = 2025
grado = np.arange(0, 30 + 1, 1) #de 0 a 100 de 1 en 1
aproxi = np.array([])
y_pred_vec = np.array([])
for i in grado:
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, anno)
    aproxi = np.append(aproxi, p)

    #Calculamos el error medio cuadratico para saber que funcion tiene menor porcentaje de error
    y_pred_vec = np.array([])
    for j in x:
      y_pred = np.polyval(coef, j)
      y_pred_vec = np.append(y_pred_vec, y_pred)

    EMC = (sum( (y - y_pred_vec)**2 ))/len(y)
    print(f"Para grado {i} el EMC es: {EMC}")


    #HACER DOS ARRAYS, UNO SE GUARDA EL NUMERO DEL GRADO Y OTRO SU EMC, SE COJE EL DEL VALOR DEL EMC MENOR Y SE MUESTRA LA GRAFICA DE ESE GRADO

    todoslosgrados = np.append(todoslosgrados, [i])
    emcdecadagrado = np.append(emcdecadagrado, [EMC])



#Dibujamos la tabla de prediccion por grado
plt.figure(figsize=[20,5])
plt.title("Grado polinomio vs predicci칩n")
plt.plot(grado,aproxi,"-",linewidth=3,color='red')
##plt.xticks(np.arange(min(grado), max(grado)+1, 1))  # Ajuste para mostrar valores en el eje x de 1 en 1
plt.grid("on")
plt.show()


#OBTENEMOS EL MENOR EMC JUNTO A SU GRADO

primeros10num = emcdecadagrado[:11]

minimoemc = min(primeros10num)
indexminimoemc = np.where(emcdecadagrado == minimoemc)[0][0]

gradodelmenoremc = todoslosgrados[indexminimoemc]

print(todoslosgrados)
print(emcdecadagrado)
print(minimoemc)
print(gradodelmenoremc)

#Hacemos la grafica del grado con mas porcentaje de prediccion

coefmejor = np.polyfit(x,y,gradodelmenoremc)
pmejor = np.polyval(coefmejor, anno)

x1 = np.linspace(1990, anno + 1, 1000)
y1 = np.polyval(coefmejor, x1)
plt.figure(figsize=[20,10])
plt.title(f"Beneficios a lo largo de los a침os con mayor tasa de prediccion en base a la funcion {gradodelmenoremc}")
plt.scatter(x,y,s=120,c='blueviolet')
plt.plot(x1,y1,"--",linewidth=3,color='orange')
plt.scatter(anno,pmejor,s=200,c='red')
plt.yticks(range(100,400,20))
plt.xticks(np.arange(1990, anno + 1, 1))  # Establecer los ticks del eje x
plt.grid(True)
ax = plt.gca()
ax.set_xlabel("$a침os$")
ax.set_ylabel("Beneficios")
plt.show()
```
