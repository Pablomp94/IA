```python
# Importamos las bibliotecas necesarias
from sklearn.tree import DecisionTreeClassifier  # Árboles de decisión
from sklearn.neighbors import KNeighborsClassifier  # K-vecinos más cercanos (KNN)
from sklearn.model_selection import train_test_split  # División del dataset en entrenamiento y prueba
from sklearn.preprocessing import LabelEncoder  # Codificar valores de texto a numéricos

# Definimos los modelos que vamos a usar
knn = KNeighborsClassifier(n_neighbors=3)  # Definimos el modelo KNN con 3 vecinos
decision_tree = DecisionTreeClassifier()  # Definimos el modelo de árbol de decisión

# Cargamos el dataset iris desde un archivo CSV
dataset = pd.read_csv('iris.csv', names=properties)  # Cargamos el archivo iris.csv, donde "properties" debe tener los nombres de las columnas

# Separar las características (X) de las etiquetas (y)
X = dataset.drop('Clase', axis=1)  # Eliminar la columna 'Clase' (etiquetas) de X, ya que solo necesitamos las características
y = dataset['Clase']  # 'Clase' contiene las etiquetas (especies de flores) que queremos predecir

# Convertir las características de texto a valores numéricos
# Usamos LabelEncoder para convertir columnas de texto en X a números (si es necesario)
le_X = LabelEncoder()
for column in X.columns:  # Recorremos cada columna de X
    if X[column].dtype == 'object':  # Si el tipo de datos de la columna es 'object' (texto)
        X[column] = le_X.fit_transform(X[column])  # Convertimos los valores de texto a números

# Codificamos la variable objetivo (y) que contiene las especies de flores
le = LabelEncoder()  # Creamos un objeto LabelEncoder
y = le.fit_transform(y)  # Convertimos las etiquetas de texto (las especies) a números

# Dividimos el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)
# train_test_split divide los datos:
# - 80% para entrenamiento (X_train, y_train)
# - 20% para prueba (X_test, y_test)
# random_state=33 asegura que la división sea reproducible

# Imprimimos cuántas instancias hay en los conjuntos de entrenamiento y prueba
print("Instancias para entrenamiento: ", X_train.shape[0])  # Debería devolver 120 instancias
print("Instancias para test: ", X_test.shape[0])  # Debería devolver 30 instancias

# Entrenamos los modelos con los datos de entrenamiento
knn.fit(X_train, y_train)  # Entrenamos el modelo KNN con el conjunto de entrenamiento
decision_tree.fit(X_train, y_train)  # Entrenamos el modelo de árbol de decisión con el conjunto de entrenamiento

# Importamos accuracy_score para medir la precisión de los modelos
from sklearn.metrics import accuracy_score

# Predecimos sobre los datos de entrenamiento para ambos modelos
prediccion = knn.predict(X_train)  # Predecimos usando el modelo KNN en los datos de entrenamiento
print('Precisión KNN en entrenamiento: ', accuracy_score(y_train, prediccion))  # Imprimimos la precisión de KNN en los datos de entrenamiento

prediccion = decision_tree.predict(X_train)  # Predecimos usando el modelo de árbol de decisión en los datos de entrenamiento
print('Precisión Árbol de Decisión en entrenamiento: ', accuracy_score(y_train, prediccion))  # Imprimimos la precisión del árbol de decisión en los datos de entrenamiento

# Predecimos sobre los datos de prueba para ambos modelos
prediccion = knn.predict(X_test)  # Predecimos usando el modelo KNN en los datos de prueba
print('Precisión KNN en prueba: ', accuracy_score(y_test, prediccion))  # Imprimimos la precisión de KNN en los datos de prueba

prediccion = decision_tree.predict(X_test)  # Predecimos usando el modelo de árbol de decisión en los datos de prueba
print('Precisión Árbol de Decisión en prueba: ', accuracy_score(y_test, prediccion))  # Imprimimos la precisión del árbol de decisión en los datos de prueba
```
